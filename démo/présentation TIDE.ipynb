```python
# 1) on charge le fichier
mob = pnd.read_csv("couverture-2g-3g-4g-en-france-par-operateur-juillet-2015.csv",sep=";", \
                   index_col=1)
mob.head()
```

```python
# 2) on construit un DataFrame avec les données des opérateurs
mob2 = pnd.concat([mob.loc[:,"Orange France Couverture population 4G":"Free Mobile Couverture population 4G"],\
                 mob.loc[:,"Orange France Couverture population 3G":"Free Mobile Couverture population 3G"],\
                 mob.loc[:,"Orange France Couverture population 2G":"Free Mobile Couverture population 2G"],\
                 mob.loc[:,"Orange France Couverture surfacique 4G":"Free Mobile Couverture surfacique 4G"],\
                 mob.loc[:,"Orange France Couverture surfacique 3G":"Free Mobile Couverture surfacique 3G"],\
                 mob.loc[:,"Orange France Couverture surfacique 2G":"Free Mobile Couverture surfacique 2G"]],
                    axis=1)
mob2.head()
```

```python
# 3) colonnes hiérarchiques Type / Techno / Opérateur
mob2.columns = [["Population"] * 12 + ["Surfacique"] * 12,\
                   ["4G"] * 4 + ["3G"] * 4 + ["2G"] * 4 + ["4G"] * 4 + ["3G"] * 4 + ["2G"] * 4,\
                    ["Orange", "Bouygues", "SFR", "Free"] * 6] 
mob2.columns.names = ["Type", "Techno", "Opérateur"]
mob2.head()
```

```python
# colonne coordonnees
mob["coordonnees"].head()
```

```python
# on insère la latitude et la longitude
mob["Latitude"] = mob["coordonnees"].apply(lambda x: float(x.split(', ')[0]))
mob["Longitude"] = mob["coordonnees"].apply(lambda x: float(x.split(', ')[1]))
```

```python
mob.head()
```

```python
# carte de France
mob.plot(kind='scatter', x="Longitude", y="Latitude");
```

```python
def carte(mob):
    # figure de la couverture 4G des 4 opérateurs
    fig = plt.figure(figsize=(14,12))
    ax1 = fig.add_subplot(221)
    var = mob.sort_values(by="Orange France Couverture surfacique 4G")
    ax1.scatter(x=var["Longitude"],\
                y=var["Latitude"],\
                c=var["Orange France Couverture surfacique 4G"],\
                cmap=plt.cm.winter,\
                edgecolors='none');
    ax1.set_title("Orange France 4G")
    ax1 = fig.add_subplot(222)
    var = mob.sort_values(by="Bouygues Telecom Couverture surfacique 4G")
    ax1.scatter(x=var["Longitude"],\
                y=var["Latitude"],\
                c=var["Bouygues Telecom Couverture surfacique 4G"],\
                cmap=plt.cm.winter,\
                edgecolors='none');
    ax1.set_title("Bouygues Telecom 4G")
    ax1 = fig.add_subplot(223)
    var = mob.sort_values(by="SFR Couverture surfacique 4G")
    ax1.scatter(x=var["Longitude"],\
                y=var["Latitude"],\
                c=var["SFR Couverture surfacique 4G"],\
                cmap=plt.cm.winter,\
                edgecolors='none');
    ax1.set_title("SFR 4G")
    ax1 = fig.add_subplot(224)
    var = mob.sort_values(by="Free Mobile Couverture surfacique 4G")
    ax1.scatter(x=var["Longitude"],\
                y=var["Latitude"],\
                c=var["Free Mobile Couverture surfacique 4G"],\
                cmap=plt.cm.winter,\
                edgecolors='none');
    ax1.set_title("Free Mobile 4G");
```

```python
carte(mob)
```
